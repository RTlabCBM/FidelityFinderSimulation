```python
#@markdown # Main program
__file__ = "main_program"

# encoding:UTF-8



########################################################################################

__doc__ = """
SYNOPSIS

  This Python script simulates an RT-PCR experiment incorporating cDNA barcoding and NGS sequencing.

  The program initiates by generating a specified number of cDNA sequences (initial_seq_number),
  each being assigned a unique barcode (also known as UMI).

  Random mutations are introduced to the cDNA sequences based on the probability defined by the RT_error_rate parameter.
  These mutations emulate transcription and reverse transcription errors.

  Subsequently, the cDNAs are then amplified through the indicated PCR cycles with a proability of introducing errors
  in the sequences (determined by PCR_error_rate) or in the barcodes associated (PCR_error_rate_for_barcodes).
  PCR_selection_rate governs the proportion of sequences amplified during each PCR cycle, thereby simulating PCR bias.

  Following PCR amplification, a subset of sequences (determined by NGS_reads_number) undergoes sequencing simulation.
  Errors may be introduced in the sequences or their associated barcodes, as defined by NGS_error_rate and NGS_error_rate_for_barcodes.
  Additionally, the program allows the introduction of errors in specific positions of the barcodes with varying probabilities,
  controlled by NGS_error_rate_for_barcodes_hotspots and modulated by hotspots_module. A hotspots_module of 2 introces errors
  in even positions of the barcodes, while a value higher than barcode_len has no effect.

  For both cDNA and barcode sequences, single nucleotide substitution errors are the sole type of simulated errors.


  Numbers are used to represent nucleotide bases in the sequences:
    0    correct nucleotide
    1    mutated nucleotide introduced in the cDNA (transcription or reverse transcription origin)
    2    mutated nucleotide introduced during PCR
    3    mutated nucleotide introduced during NGS

  For barcode sequences, the same numerical system is employed, but the numbers hold no inherent meaning.

  This script is an adaptation of the original program to be run on Google Colab.


  AUTHORS

    Javier Martinez del Río (javier.martinez@cbm.csic.es; javier.mardelrio@gmail.com)

"""

__version__ = 'v1.0.0'




## Imports
##----------

import numpy as np
import random
from tqdm import tqdm
import matplotlib.pyplot as plt
import sys, os
from collections import defaultdict
from collections import Counter
from scipy import stats
import pandas as pd
import json





## Functions
##----------
def send_error_message(message):
    """Gives an error message"""
    print("\n\t########## ERROR ########## \n{}\n".format(message))
    sys.exit(1)

def parse_list(input_numbers):
    """
    Parses the input string or tuple and returns a list of integers.

    Parameters:
    - input_string (str/tuple): Input string containing values separated by commas.

    Returns:
    - list of int: List of values as integers.
    """
    input_numbers = str(input_numbers)
    if (input_numbers.startswith("(") and input_numbers.endswith(")") or
        input_numbers.startswith("[") and input_numbers.endswith("]")):
        input_numbers = input_numbers[1:-1]
    try:
        # Try to convert the input string into a list of numbers
        output_list = [int(val) for val in input_numbers.split(',')]
        return output_list
    except ValueError:
        raise send_error_message(f"Invalid threshold format: {input_numbers}. Please provide a comma-separated list of integers.")

def save_parameters_to_file(filename, **kwargs):
    """
    Saves the parameters and their values to a JSON file.

    Parameters:
        filename (str): Name of the file to save the parameters.
        **kwargs: Keyword arguments, where the name of each parameter is the key and its value is the value.

    Returns:
        str: Name of the file where the parameters were saved.
    """
    with open(filename, "w") as file:
        json.dump(kwargs, file)
    return filename

def generate_random_base_four(digit_number):
    """
    Generates a random base-4 representation with a fixed length specified in digit_number.

    Parameters:
    - digit_number (int): The length of the base-4 representation.

    Returns:
    - random_base_four (str): A random base-4 representation with the specified length.
    """
    # Generate a random number in base-4 representation
    random_number = ''.join(str(random.randint(0, 3)) for _ in range(digit_number))
    return random_number

def count_string_differences(str1, str2):
    """
    Count the number of differing characters between two strings at each position.

    Parameters:
    - str1 (str): The first input string.
    - str2 (str): The second input string.

    Returns:
    - int: The total number of differences between the two strings.
    """
    # Ensure both strings have the same length
    if len(str1) != len(str2):
        raise ValueError("Input strings must have the same length.")

    # Count the number of differing characters at each position
    differences = sum(1 for char1, char2 in zip(str1, str2) if char1 != char2)

    return differences

def calculate_frequencies_from_list(numbers_list):
    # Use Counter to calculate the frequency of each number
    numbers_frequency = Counter(numbers_list)

    # Sort the frequency of numbers by the values of the keys
    sorted_frequency = sorted(numbers_frequency.items())

    return sorted_frequency


def summary_mutation_statistics(sequences_list, seq_len):
  total_mutations = 0
  total_RT_mutations = 0
  total_PCR_mutations = 0
  total_NGS_mutations = 0
  offspring_barcodes = 0
  PCR_offspring_barcodes = 0
  NGS_offspring_barcodes = 0
  PCR_and_NGS_offspring_barcodes = 0
  offspring_barcodes_differences = []
  for sequence_dic in sequences_list:
    total_RT_mutations += sequence_dic['sequence'].count('1')
    total_PCR_mutations += sequence_dic['sequence'].count('2')
    total_NGS_mutations += sequence_dic['sequence'].count('3')
    total_mutations = total_RT_mutations + total_PCR_mutations + total_NGS_mutations
    original_barcode = sequence_dic['original_barcode']
    resulting_barcode = sequence_dic['resulting_barcode']
    barcode_mutations_list = sequence_dic['barcode_mutations_list']
    if original_barcode != resulting_barcode:
      offspring_barcodes += 1
      number_of_differences = count_string_differences(original_barcode, resulting_barcode)
      offspring_barcodes_differences.append(number_of_differences)
      if barcode_mutations_list[0] > 0 and barcode_mutations_list[1]==0:
        PCR_offspring_barcodes += 1
      elif barcode_mutations_list[0] == 0 and barcode_mutations_list[1] > 0:
        NGS_offspring_barcodes += 1
      elif barcode_mutations_list[0] > 0 and barcode_mutations_list[1] > 0:
        PCR_and_NGS_offspring_barcodes += 1

  total_sequences = len(sequences_list)
  error_rate = total_mutations/(total_sequences * seq_len)
  RT_error_rate = total_RT_mutations/(total_sequences * seq_len)
  PCR_error_rate = total_PCR_mutations/(total_sequences * seq_len)
  NGS_error_rate = total_NGS_mutations/(total_sequences * seq_len)
  offsprings_differences_frequencies = calculate_frequencies_from_list(offspring_barcodes_differences)

  print("\n")
  print("Summary mutation statistics")
  print("\tTotal sequences: ", total_sequences)
  print("\tTotal mutations: ", total_mutations)
  print("\tTotal RT mutations: ", total_RT_mutations)
  print("\tTotal PCR mutations: ", total_PCR_mutations)
  print("\tTotal NGS mutations: ", total_NGS_mutations)
  print("\tError rate: ", error_rate)
  print("\tRT error rate: ", RT_error_rate)
  print("\tPCR error rate: ", PCR_error_rate)
  print("\tNGS error rate: ", NGS_error_rate)
  print("\tTotal barcodes: ", total_sequences)
  print("\tOffspring barcodes: ", offspring_barcodes)
  print("\tPCR offspring barcodes: ", PCR_offspring_barcodes)
  print("\tNGS offspring barcodes: ", NGS_offspring_barcodes)
  print("\tPCR and NGS offspring barcodes: ", PCR_and_NGS_offspring_barcodes)
  print("\tOfffsprings differences frequencies: ", offsprings_differences_frequencies)
  print("\n")


def generate_cdnas(initial_seq_number, seq_len, barcode_len, RT_error_rate):
    """
    Generates a list of cDNA sequences with barcodes

    Parameters:
    - initial_seq_number (int): Number of cDNA sequences to be generated.
    - seq_len (int): Length of the sequences (nt).
    - barcode_len (int): Length of the barcode added to each sequence (nt)
    - RT_error_rate (float): Reverse transcription error rate.

    Returns:
    - sequences_list (list): List of dictionaries representing cDNA sequences with barcodes.
    """
    sequences_list = []
    for i in range(initial_seq_number):
        seq_dic = {}
        # Generate and assign original and resulting barcodes
        barcode = generate_random_base_four(digit_number=barcode_len)
        seq_dic["original_barcode"] = barcode
        seq_dic["resulting_barcode"] = barcode
        seq_dic["barcode_mutations_list"] = [0,0]

        # Generate a random sequence based on the reverse transcription error rate
        sequence = ""
        for nt in range(seq_len):
            if random.random() > RT_error_rate:
                sequence += "0"
            else:
                sequence += "1"
        seq_dic["sequence"] = sequence

        sequences_list.append(seq_dic)

    return sequences_list



def perform_PCR(sequences_list, PCR_cycles_number, PCR_error_rate, PCR_selection_rate):
  for cycle in tqdm(range(PCR_cycles_number)):
    print("PCR cycle: ", cycle+1)

    # Select sequences to amplify based on PCR_selection_rate parameter
    number_sequences_to_amplify = int(PCR_selection_rate * len(sequences_list))
    sequences_to_amplify = random.sample(sequences_list, number_sequences_to_amplify)

    # Amplify selected sequences (sequences_to_amplify) taking into account the PCR_error_rate
    amplified_sequences = []
    for sequence_dic in sequences_to_amplify:

      original_barcode = sequence_dic['original_barcode']
      barcode = sequence_dic['resulting_barcode']
      sequence = sequence_dic['sequence']
      barcode_mutations_list = sequence_dic['barcode_mutations_list'][:]
      new_sequence = ""
      new_barcode = ""
      for nt in sequence:
        if random.random() > PCR_error_rate:
            new_sequence += nt
        else:
            new_sequence += "2"
      for nt in barcode:
        if random.random() > PCR_error_rate_for_barcodes:
            new_barcode += nt
        else:
            new_barcode += np.random.choice([x for x in ['0','1','2','3'] if x != nt])
            barcode_mutations_list[0] += 1
      new_seq_dict= {"original_barcode": original_barcode, "resulting_barcode": new_barcode, "barcode_mutations_list": barcode_mutations_list, "sequence": new_sequence}
      amplified_sequences.append(new_seq_dict)

    # Combine the initial sequences_list and amplified_sequences
    sequences_list = sequences_list + amplified_sequences
    summary_mutation_statistics(sequences_list, seq_len)

  return sequences_list

def perform_NGS(sequences_list, NGS_reads_number, NGS_error_rate):
  if NGS_reads_number > len(sequences_list):
    NGS_reads_number = len(sequences_list)
    print(f"WARNING: NGS_reads_number is higher than the number of sequences obtained by PCR ({len(sequences_list)})")
  sequences_to_NGS = random.sample(sequences_list, NGS_reads_number)
  sequenced_sequences = []
  for sequence_dic in sequences_to_NGS:
    original_barcode = sequence_dic['original_barcode']
    barcode = sequence_dic['resulting_barcode']
    sequence = sequence_dic['sequence']
    barcode_mutations_list = sequence_dic['barcode_mutations_list'][:]
    new_sequence = ""
    new_barcode = ""
    for nt in sequence:
      if random.random() > NGS_error_rate:
          new_sequence += nt
      else:
          new_sequence += "3"
    for barcode_position,nt in enumerate(barcode):
    # During barcode sequencing, the error rate varies based on position.
    # If the position % hotspots_module == 0 and nt == 0, a sequencing error occurs with a probability defined by 'NGS_error_rate_for_barcodes_hotspots'
    # and mutation introduced is always the same one (nt: 1)
    # Otherwise, errors occur with a probability defined by 'NGS_error_rate'.
      if barcode_position % hotspots_module == 0 and nt == '0':
        if random.random() > NGS_error_rate_for_barcodes_hotspots:
            new_barcode += nt
        else:
            new_barcode += '1'
            barcode_mutations_list[1] += 1
      else:
        if random.random() > NGS_error_rate_for_barcodes:
            new_barcode += nt
        else:
            new_barcode += np.random.choice([x for x in ['0','1','2','3'] if x != nt])
            barcode_mutations_list[1] += 1
    new_seq_dict= {"original_barcode": original_barcode, "resulting_barcode": new_barcode, "sequence": new_sequence, "barcode_mutations_list": barcode_mutations_list}
    sequenced_sequences.append(new_seq_dict)
  return sequenced_sequences


def create_fasta_dics(sequences_list):
  """
  Create dictionary with barcodes as keys and list with their associated sequences
  """
  fasta_dic = defaultdict(list)
  offsprings_fasta_dic =  defaultdict(list)

  for sequence_dic in tqdm(sequences_list):

    original_barcode = sequence_dic['original_barcode']
    barcode = sequence_dic['resulting_barcode']
    sequence = sequence_dic['sequence']

    fasta_dic[barcode].append(sequence)

    if barcode != original_barcode:
      offsprings_fasta_dic[barcode].append(sequence)

  return fasta_dic, offsprings_fasta_dic


def create_barcodes_dic(fasta_dic):

    barcodes_dic = defaultdict(int)
    for barcode in tqdm(fasta_dic.keys()):
      seqList  = fasta_dic[barcode]
      nseqs    = len(seqList)
      barcodes_dic[barcode] = nseqs

    sorted_barcodes_dic = dict(sorted(barcodes_dic.items()))
    return sorted_barcodes_dic


def calculate_frequency_of_frequencies(fasta_dic):
    """
    Calculate the frequencies of frequencies in the given dictionary.

    Parameters:
        fasta_dic (dict): Dictionary containing barcodes as keys and values are a list of sequences with each barcodes

    Returns:
        sorted_result_dict (dict): A dictionary where keys are frequencies and values are the count of occurrences of each frequency.
    """
    counted_barcodes_frequencies_dic = defaultdict(int)
    for barcode in tqdm(fasta_dic.keys()):
      seqList  = fasta_dic[barcode]
      nseqs    = len(seqList)
      counted_barcodes_frequencies_dic[nseqs] += 1
    sorted_counted_barcodes_frequencies_dic = dict(sorted(counted_barcodes_frequencies_dic.items()))
    return sorted_counted_barcodes_frequencies_dic


def draw_graph(output_prefix, counted_barcodes_frequencies_dic, y_log_scale=True):
    """
    Draw a graph for primer ID distribution.

    Parameters:
    - output_prefix (str): Prefix for the output files.
    - counted_barcodes_frequencies_dic (dict): Dictionary with barcode frequencies.
    """
    x = counted_barcodes_frequencies_dic.keys()
    y = counted_barcodes_frequencies_dic.values()
    plt.scatter(x, y, marker="o")
    if y_log_scale == True:
      plt.yscale('log')
    titlename = 'Barcodes frequencies distribution for ' + output_prefix
    plt.title(titlename)
    plt.xlabel('Barcodes frequencies')
    plt.ylabel('Occurrences')

    # Save the plot
    frequencies_graph = output_prefix + "_frequencies_distribution_graph.png"
    plt.savefig(frequencies_graph)
    plt.show()
    plt.close()

    return frequencies_graph

def draw_combined_graph(output_prefix, counted_barcodes_frequencies_dic, counted_offsprings_barcodes_frequencies_dic, x_log_scale=False, y_log_scale=False):
    """
    Draw a graph for primer ID distribution.

    Parameters:
    - output_prefix (str): Prefix for the output files.
    - counted_barcodes_frequencies_dic (dict): Dictionary with barcode frequencies.
    """
    plt.scatter(counted_barcodes_frequencies_dic.keys(), counted_barcodes_frequencies_dic.values(), label='counted_barcodes_frequencies_dic', marker='o')
    plt.scatter(counted_offsprings_barcodes_frequencies_dic.keys(), counted_offsprings_barcodes_frequencies_dic.values(), label='counted_offsprings_barcodes_frequencies_dic', marker='x')
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

    if x_log_scale == True:
      plt.xscale('log')

    if y_log_scale == True:
      plt.yscale('log')

    titlename = 'Barcodes frequencies distribution for ' + output_prefix
    plt.title(titlename)
    plt.xlabel('Barcodes frequencies')
    plt.ylabel('Occurrences')

    # Save the plot
    combined_frequencies_graph = output_prefix + "_frequencies_distribution_graph.png"
    plt.savefig(combined_frequencies_graph, bbox_inches='tight')
    plt.show()
    plt.close()
    return combined_frequencies_graph


def draw_combined_percentages_graph(output_prefix, counted_barcodes_frequencies_dic, counted_offsprings_barcodes_frequencies_dic, x_log_scale=False, y_log_scale=False):
    """
    Draw a graph for primer ID distribution.

    Parameters:
    - output_prefix (str): Prefix for the output files.
    - counted_barcodes_frequencies_dic (dict): Dictionary with barcode frequencies.
    """
    # Extract frequencies and occurrences/percentages for each dataset
    frequencies1, occurrences1 = list(counted_barcodes_frequencies_dic.keys()), list(counted_barcodes_frequencies_dic.values())

    percentage_offsprings_1differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_barcodes_frequencies_dic.items()}
    frequencies2, percentages2 = list(percentage_offsprings_1differenc_frequencies_dic.keys()), list(percentage_offsprings_1differenc_frequencies_dic.values())

    # Create a scatter plot
    fig, ax = plt.subplots()
    ax2 = ax.twinx()
    ax2.set_ylim(0, 101)
    ax.scatter(frequencies1, occurrences1, color='tab:blue', label='counted_barcodes_frequencies_dic', marker='o')
    ax2.plot(frequencies2, percentages2, '^-', color='tab:orange', label='% counted_offsprings_barcodes_frequencies_dic')

    # Add labels and title
    ax.set_xlabel('Barcodes frequencies')
    ax.set_ylabel('Occurrences')
    ax2.set_ylabel('% of offspring barcodes')
    ax.set_title('Barcodes frequencies distribution for ' + output_prefix)

    # Set axis to log scale
    if x_log_scale == True:
      plt.xscale('log')

    if y_log_scale == True:
      plt.yscale('log')

    # Add legends
    ax.legend(loc='center left', bbox_to_anchor=(1.2, 0.75))
    ax2.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))

    # Save the plot
    combined_frequencies_percentages_graph = output_prefix + "_frequencies_percentages_distribution_graph.png"
    plt.savefig(combined_frequencies_percentages_graph, bbox_inches='tight')
    plt.show()
    plt.close()
    return combined_frequencies_percentages_graph

def create_histogram(data_list, output_prefix, x_label):
    # Histogram configuration
    bin_size = 0.5
    bins = [i * bin_size for i in range(int(100 / bin_size) + 1)]
    plt.hist(data_list, bins=bins)

    # Plot configuration
    plt.title(f'{output_prefix} histogram')
    plt.xlabel(x_label)
    plt.ylabel('Frequence')
    plt.yscale('log')
    plt.grid(True)

    # Save the plot
    histogram_graph = output_prefix + '_histogram.png'
    plt.savefig(histogram_graph)
    plt.show()
    plt.close()
    return histogram_graph

def create_percetages_histogram(cutoff, threshold, data_list, output_prefix, x_label):
    # Histogram configuration
    bin_size = 0.5
    bins = [i * bin_size for i in range(int(100 / bin_size) + 1)]
    plt.hist(data_list, bins=bins)

    # Plot configuration
    plt.title(f'{output_prefix} histogram')
    plt.xlabel(x_label)
    plt.ylabel('Frequence')
    plt.yscale('log')
    plt.grid(True)

    # Add text with the values of cutoff and threshold
    plt.text(0.5, 0.95, f'Cutoff: {cutoff}', transform=plt.gca().transAxes, ha='center', va='center')
    plt.text(0.5, 0.9, f'Threshold: {threshold}', transform=plt.gca().transAxes, ha='center', va='center')

    # Save the plot
    histogram_graph = f"{output_prefix}_percentages_histogram_cutoff_{cutoff}_threshold_{threshold}.png"
    plt.savefig(histogram_graph)
    plt.show()
    plt.close()
    return histogram_graph

def find_most_frequent_element(data_list, threshold):
    counter = Counter(data_list)

    # Get the element(s) with the highest frequency
    max_frequency = max(counter.values())
    elements_with_max_frequency = [element for element, freq in counter.items() if freq == max_frequency]

    # Check if no element surpasses the threshold
    if max_frequency / len(data_list) * 100 < threshold:
        return 'N'

    # Check if there is more than one element with the highest frequency
    if len(elements_with_max_frequency) > 1:
        return 'N'

    # Return the element(s) with the highest frequency
    return elements_with_max_frequency[0]


def construct_consensus(sequences_list, threshold):
    positions_dic = defaultdict(list)
    for sequence in sequences_list:
        for position, nucleotide in enumerate(sequence):
            positions_dic[position].append(nucleotide)
    consensus = ''.join(find_most_frequent_element(nucleotide_list, threshold) for nucleotide_list in positions_dic.values())
    return consensus, positions_dic

def count_nucleotides_percentage(positions_dic):
  max_frequency_percentages = []
  RT_mutations_percentages = []
  PCR_mutations_percentages = []
  NGS_mutations_percentages = []

  for position, nucleotide_list in positions_dic.items():
    counter = Counter(nucleotide_list)
    total_number_nucleotides = len(nucleotide_list)

    max_frequency_percentage = round(max(counter.values())/total_number_nucleotides * 100, 1)
    RT_mutations_percentage = round(nucleotide_list.count("1")/total_number_nucleotides * 100, 1)
    PCR_mutations_percentage = round(nucleotide_list.count("2")/total_number_nucleotides * 100, 1)
    NGS_mutations_percentage = round(nucleotide_list.count("3")/total_number_nucleotides * 100, 1)

    max_frequency_percentages.append(max_frequency_percentage)
    RT_mutations_percentages.append(RT_mutations_percentage)
    PCR_mutations_percentages.append(PCR_mutations_percentage)
    NGS_mutations_percentages.append(NGS_mutations_percentage)

  return max_frequency_percentages, RT_mutations_percentages, PCR_mutations_percentages, NGS_mutations_percentages

def calculate_error_rate(mutations, seq_len, total_consensus_sequences):
    try:
        error_rate = mutations / (seq_len * total_consensus_sequences)
    except ZeroDivisionError:
        error_rate = np.NaN
    return error_rate


def count_observed_mutations(consensus_dic, seq_len, cutoff, threshold, RT_error_rate):

  total_consensus_sequences = len(consensus_dic)

  observed_RT_mutations = sum(sequence.count('1') for sequence in consensus_dic.values())
  observed_PCR_mutations = sum(sequence.count('2') for sequence in consensus_dic.values())
  observed_NGS_mutations = sum(sequence.count('3') for sequence in consensus_dic.values())
  observed_total_mutations = observed_RT_mutations + observed_PCR_mutations + observed_NGS_mutations

  observed_error_rate = calculate_error_rate(observed_total_mutations, seq_len, total_consensus_sequences)
  observed_RT_error_rate = calculate_error_rate(observed_RT_mutations, seq_len, total_consensus_sequences)
  observed_PCR_error_rate = calculate_error_rate(observed_PCR_mutations, seq_len, total_consensus_sequences)
  observed_NGS_error_rate = calculate_error_rate(observed_NGS_mutations, seq_len, total_consensus_sequences)

  error_rate_accuracy = observed_error_rate / RT_error_rate * 100 if RT_error_rate != 0 else float('nan')

  print(f'\t\tTotal consensus sequences constructed using a cutoff of {cutoff} and a threshold of {threshold}%: {total_consensus_sequences}')
  print(f'\t\tTotal mutations observed: {observed_total_mutations}')
  print(f'\t\tRT mutations: {observed_RT_mutations}')
  print(f'\t\tPCR mutations: {observed_PCR_mutations}')
  print(f'\t\tNGS mutations: {observed_NGS_mutations}')
  print(f'\t\tError rate: {observed_error_rate}')
  print(f'\t\tRT error rate accuracy (observed_error_rate / RT_error_rate * 100 ): {error_rate_accuracy}')
  print(f'')

  summary_data = {
    'cutoff': cutoff,
    'threshold': threshold,
    'consensus': total_consensus_sequences,
    'seq_len': seq_len,
    'expected_RT_error_rate': RT_error_rate,
    'total_mutations': observed_total_mutations,
    'observed_RT_mutations': observed_RT_mutations,
    'observed_PCR_mutations': observed_PCR_mutations,
    'observed_NGS_mutations': observed_NGS_mutations,
    'error_rate_accuracy': error_rate_accuracy,
    'observed_error_rate': observed_error_rate,
    'observed_RT_error_rate': observed_RT_error_rate,
    'observed_PCR_error_rate': observed_PCR_error_rate,
    'observed_NGS_error_rate': observed_NGS_error_rate
  }

  return summary_data


def count_mutations_positions(consensus_dic):
    try:
      rt_mutations_positions = {position: 0 for position in range(len(next(iter(consensus_dic.values()))))}
      pcr_mutations_positions = {position: 0 for position in range(len(next(iter(consensus_dic.values()))))}
      ngs_mutations_positions = {position: 0 for position in range(len(next(iter(consensus_dic.values()))))}

      for barcode, consensus_sequence in consensus_dic.items():
          for position, nt in enumerate(consensus_sequence):
              if nt == '1':
                  rt_mutations_positions[position] += 1
              elif nt == '2':
                  pcr_mutations_positions[position] += 1
              elif nt == '3':
                  ngs_mutations_positions[position] += 1
      return rt_mutations_positions, pcr_mutations_positions, ngs_mutations_positions

    except Exception as e:
        print(f"WARNING: An error occurred - {str(e)}")
        return {}, {}, {}

def plot_mutations_bar_graph(output_prefix, cutoff, threshold, rt_mutations, pcr_mutations, ngs_mutations):
    positions = list(rt_mutations.keys())
    rt_counts = list(rt_mutations.values())
    pcr_counts = list(pcr_mutations.values())
    ngs_counts = list(ngs_mutations.values())

    width = 0.25  # Width of the bars

    plt.bar(positions, rt_counts, width, align='center', alpha=0.7, label='RT Mutations')
    plt.bar(positions, pcr_counts, width, align='edge', alpha=0.7, label='PCR Mutations')
    plt.bar(positions, ngs_counts, width, align='edge', alpha=0.7, label='NGS Mutations')

    plt.xlabel('Sequence positions')
    plt.ylabel('Mutation counts')
    plt.title('Mutations distribution')
    plt.legend()

    # Add text with the values of cutoff and threshold
    plt.text(0.5, 0.95, f'Cutoff: {cutoff}', transform=plt.gca().transAxes, ha='center', va='center')
    plt.text(0.5, 0.9, f'Threshold: {threshold}', transform=plt.gca().transAxes, ha='center', va='center')

    # Save the plot
    mutations_bar_graph = f"{output_prefix}mutations_distribution_graph_cutoff_{cutoff}_threshold_{threshold}.png"
    plt.savefig(mutations_bar_graph)
    plt.show()
    plt.close()
    return mutations_bar_graph

def plot_error_rate(output_prefix, summary_df, RT_error_rate, threshold, ylim=None):
    # Filter the data according to the specified criteria
    filtered_df = summary_df[summary_df['threshold'] == threshold]

    # Create the plot
    plt.plot(filtered_df['cutoff'], filtered_df['observed_error_rate'], marker='o', label='Observed Error Rate')

    plt.xlabel('Cutoff')
    plt.ylabel('Error Rate')
    plt.title(f'Error Rate vs Cutoff (Threshold {threshold})')
    plt.grid(True)

    # Set y limit if provided
    if ylim is not None:
        plt.ylim(ylim)

    # Add a horizontal line with the expected error rate
    plt.axhline(y=RT_error_rate, color='r', linestyle='--', label='Expected RT error Rate')
    plt.legend()

    # Save the plot
    error_rates_graph = output_prefix+"error_rates_graph.png"
    plt.savefig(error_rates_graph)
    plt.show()
    plt.close()
    return error_rates_graph

def plot_all_error_rates(output_prefix, summary_df, RT_error_rate, threshold, ylim=None):
    # Filter the data according to the specified criteria
    filtered_df = summary_df[summary_df['threshold'] == threshold]

    # Create the plot
    plt.plot(filtered_df['cutoff'], filtered_df['observed_error_rate'], marker='o', label='Observed Error Rate')
    plt.plot(filtered_df['cutoff'], filtered_df['observed_RT_error_rate'], marker='s', label='Observed RT Error Rate')
    plt.plot(filtered_df['cutoff'], filtered_df['observed_PCR_error_rate'], marker='^', label='Observed PCR Error Rate')
    plt.plot(filtered_df['cutoff'], filtered_df['observed_NGS_error_rate'], marker='d', label='Observed NGS Error Rate')

    plt.xlabel('Cutoff')
    plt.ylabel('Error Rate')
    plt.title(f'Error Rate vs Cutoff (Threshold {threshold})')
    plt.grid(True)

    # Set y limit if provided
    if ylim is not None:
        plt.ylim(ylim)

    # Add a horizontal line with the expected error rate
    plt.axhline(y=RT_error_rate, color='r', linestyle='--', label='Expected RT error Rate')
    plt.legend()

    # Save the plot
    all_error_rates_graph = output_prefix+"all_error_rates_graph.png"
    plt.savefig(all_error_rates_graph)
    plt.show()
    plt.close()
    return all_error_rates_graph





## Main program
##-------------


## Step 1: Parameters catching
##--------------------

print("Step 1: Parameters catching")


# Assert inputs
assert isinstance(initial_seq_number, (int, float)), "initial_seq_number should be a number"
assert isinstance(seq_len, int) and seq_len > 0, "seq_len should be a positive integer"
assert isinstance(barcode_len, int) and barcode_len > 0, "barcode_len should be a positive integer"
assert isinstance(RT_error_rate, (int, float)) and 0 <= RT_error_rate <= 1, "RT_error_rate should be between 0 and 1"
assert isinstance(PCR_cycles_number, int) and PCR_cycles_number > 0, "PCR_cycles_number should be a positive integer"
assert isinstance(PCR_selection_rate, (int, float)) and 0 <= PCR_selection_rate <= 1, "PCR_selection_rate should be between 0 and 1"
assert isinstance(PCR_error_rate, (int, float)) and 0 <= PCR_error_rate <= 1, "PCR_error_rate should be between 0 and 1"
assert isinstance(PCR_error_rate_for_barcodes, (int, float)) and 0 <= PCR_error_rate <= 1, "PCR_error_rate_for_barcodes should be between 0 and 1"
assert isinstance(NGS_reads_number, int) and NGS_reads_number > 0, "NGS_reads_number should be a positive integer"
assert isinstance(NGS_error_rate, (int, float)) and 0 <= NGS_error_rate <= 1, "NGS_error_rate should be between 0 and 1"
assert isinstance(NGS_error_rate_for_barcodes, (int, float)) and 0 <= NGS_error_rate_for_barcodes <= 1, "NGS_error_rate_for_barcodes should be between 0 and 1"
assert isinstance(NGS_error_rate_for_barcodes_hotspots, (int, float)) and 0 <= NGS_error_rate_for_barcodes_hotspots <= 1, "NGS_error_rate_for_barcodes_hotspots should be between 0 and 1"
assert isinstance(hotspots_module, (int)) and 1 <= hotspots_module, "hotspots_module should be an integer greater or equal to 1"

cutoffs_list = parse_list(cutoffs_list)
thresholds_list = parse_list(thresholds_list)

input_parameters_file = save_parameters_to_file(output_prefix+"input_parameters.json",
                        initial_seq_number=initial_seq_number,
                        seq_len=seq_len,
                        barcode_len=barcode_len,
                        RT_error_rate=RT_error_rate,
                        PCR_cycles_number=PCR_cycles_number,
                        PCR_selection_rate=PCR_selection_rate,
                        PCR_error_rate=PCR_error_rate,
                        PCR_error_rate_for_barcodes=PCR_error_rate_for_barcodes,
                        NGS_reads_number=NGS_reads_number,
                        NGS_error_rate=NGS_error_rate,
                        NGS_error_rate_for_barcodes=NGS_error_rate_for_barcodes,
                        NGS_error_rate_for_barcodes_hotspots=NGS_error_rate_for_barcodes_hotspots,
                        hotspots_module=hotspots_module,
                        cutoffs_list=cutoffs_list,
                        thresholds_list=thresholds_list)


print("\n")
print("INITIAL PARAMETERS")
print("\tNumber of cDNA sequences that are generated by reverse transciption", initial_seq_number)
print("\tLenght of the initial_sequences (nt):", seq_len)
print("\tLength of the barcode added to each sequence (nt):", barcode_len)
print("\tError rate during reverse transcription:", RT_error_rate)
print("\tNumber of PCR cycles:", PCR_cycles_number)
print("\tProportion of sequences that are amplified in each PCR cycle (0.0-1.0):", PCR_selection_rate)
print("\tError rate during each cycle of PCR:", PCR_error_rate)
print("\tError rate during each cycle of PCR for barcode amplification:", PCR_error_rate_for_barcodes)
print("\tNumber of reads that are sequenced:", NGS_reads_number)
print("\tError rate during sequencing:", NGS_error_rate)
print("\tError rate during sequencing of barcodes:", NGS_error_rate_for_barcodes)
print("\tError rate during sequencing of barcodes hotspots:", NGS_error_rate_for_barcodes_hotspots)
print("\tHotspots module:", hotspots_module)
print("\tCutoffs for sequence filtering:", cutoffs_list)
print("\tThresholds for consensus construction:", thresholds_list)
print("\tOutputn prefix for files:", output_prefix)
print("\n")


## Step 2: Reverse transcription
##--------------------

print("Step 2: Reverse transcription")

cdna_sequences = generate_cdnas(initial_seq_number, seq_len, barcode_len, RT_error_rate)
summary_mutation_statistics(cdna_sequences, seq_len)


## Step 3: PCR
##--------------------

print("Step 3: PCR")

amplified_sequences = perform_PCR(sequences_list=cdna_sequences, PCR_cycles_number=PCR_cycles_number, PCR_error_rate=PCR_error_rate, PCR_selection_rate=PCR_selection_rate)


## Step 4: NGS
##--------------------

print("Step 4: NGS")

sequenced_sequences = perform_NGS(sequences_list=amplified_sequences, NGS_reads_number=NGS_reads_number, NGS_error_rate=NGS_error_rate)
summary_mutation_statistics(sequenced_sequences, seq_len)


## Step 5: Create fasta_dics
##--------------------

print("Step 5: Create fasta_dics")

fasta_dic, offsprings_fasta_dic = create_fasta_dics(sequenced_sequences)


## Step 6: Calculating number of occurrences for offspring barcodes frequencies
##--------------------

print("Step 6: Calculating number of occurrences for offspring barcodes frequencies")

barcodes_dic = create_barcodes_dic(fasta_dic)

counted_barcodes_frequencies_dic = calculate_frequency_of_frequencies(fasta_dic)
counted_offsprings_barcodes_frequencies_dic = calculate_frequency_of_frequencies(offsprings_fasta_dic)

barcodes_frequencies_file = save_parameters_to_file(output_prefix+"barcodes_frequencies.json",
                        counted_barcodes_frequencies_dic=counted_barcodes_frequencies_dic,
                        counted_offsprings_barcodes_frequencies_dic=counted_offsprings_barcodes_frequencies_dic)

print("Barcodes counted frequencies: ", counted_barcodes_frequencies_dic)
print("Offspring barcodes counted frequencies: ", counted_offsprings_barcodes_frequencies_dic)


## Step 7: Draw graphs
##--------------------

print("Step 7: Draw graphs")

# Barcodes and offspring barcodes frequencies distribution graphs
barcodes_frequencies_graph = draw_graph(output_prefix+"total", counted_barcodes_frequencies_dic, False)
offsprings_frequencies_graph = draw_graph(output_prefix+"offsprings", counted_offsprings_barcodes_frequencies_dic, False)
barcodes_frequencies_log_graph = draw_graph(output_prefix+"total_log", counted_barcodes_frequencies_dic, True)
offsprings_frequencies_log_graph = draw_graph(output_prefix+"offsprings_log", counted_offsprings_barcodes_frequencies_dic, True)


# Combined graphs
combined_frequencies_logx_graph = draw_combined_graph(output_prefix+"combined",
                                                counted_barcodes_frequencies_dic,
                                                counted_offsprings_barcodes_frequencies_dic,
                                                True,
                                                False)
combined_frequencies_percentages_logx_graph = draw_combined_percentages_graph(output_prefix+"combined",
                                                counted_barcodes_frequencies_dic,
                                                counted_offsprings_barcodes_frequencies_dic,
                                                True,
                                                False)

# Histograms
barcode_PCR_mutations_values = [entry['barcode_mutations_list'][0] for entry in sequenced_sequences]
barcode_NGS_mutations_values = [entry['barcode_mutations_list'][1] for entry in sequenced_sequences]
PCR_mutations_histogram = create_histogram(barcode_PCR_mutations_values,
                                           output_prefix + "PCR mutations per barcode", 'Number of PCR mutations per barcode')
NGS_mutations_histogram = create_histogram(barcode_NGS_mutations_values,
                                           output_prefix + "NGS mutations per barcode", 'Number of NGS mutations per barcode')



## Step 8: Create consensus sequences and calculate error rate
##--------------------

print("Step 8: Create consensus sequences and calculate error rate")

summary_mutations_data_list = []

#Calculate number of barcodes for different cutoff_values
print("Barcodes counted frequencies: ", counted_barcodes_frequencies_dic)

#Calculate max barcode frequency
max_barcode_frequency = max(counted_barcodes_frequencies_dic.keys())



for cutoff in cutoffs_list:
  print(f'Cutoff: {cutoff}')
  for threshold in thresholds_list:
    print(f'\tThreshold: {threshold}')
    consensus_dic = {}
    max_frequency_percentage_list = []
    RT_mutations_percentage_list = []
    PCR_mutations_percentage_list = []
    NGS_mutations_percentage_list = []
    for barcode, seqList in tqdm(fasta_dic.items()):
        nseqs = len(seqList)
        if nseqs > cutoff:
          consensus_sequence, positions_dic = construct_consensus(seqList, threshold)
          consensus_dic[barcode] = consensus_sequence

          max_frequency_percentages, RT_mutations_percentages, PCR_mutations_percentages, NGS_mutations_percentages = count_nucleotides_percentage(positions_dic)
          max_frequency_percentage_list.extend(max_frequency_percentages)
          RT_mutations_percentage_list.extend(RT_mutations_percentages)
          PCR_mutations_percentage_list.extend(PCR_mutations_percentages)
          NGS_mutations_percentage_list.extend(NGS_mutations_percentages)

    # Plot histograms

      # Histograms
    if threshold == thresholds_list[0]:
      #threshold does not affect these histograms, thus only they are only created for one threshold value
      max_frequency_percentage_histogram = create_percetages_histogram(cutoff, threshold, max_frequency_percentage_list, output_prefix+"Max_frequent_nucleotides", 'nucleotide %')
      RT_mutations_percentage_histogram = create_percetages_histogram(cutoff, threshold, RT_mutations_percentage_list, output_prefix+"RT_mutations_nucleotides", 'nucleotide %')
      PCR_mutations_percentage_histogram = create_percetages_histogram(cutoff, threshold, PCR_mutations_percentage_list, output_prefix+"PCR_mutations_nucleotides", 'nucleotide %')
      NGS_mutations_percentage_histogram = create_percetages_histogram(cutoff, threshold, NGS_mutations_percentage_list, output_prefix+"NGS_mutations_nucleotides", 'nucleotide %')

      # Plot mutation counts vs sequence position
    rt_mutations_positions, pcr_mutations_positions, ngs_mutations_positions = count_mutations_positions(consensus_dic)
    mutations_bar_graph = plot_mutations_bar_graph(output_prefix, cutoff, threshold, rt_mutations_positions, pcr_mutations_positions, ngs_mutations_positions)

    # Print statistics
    summary_data = count_observed_mutations(consensus_dic, seq_len, cutoff, threshold, RT_error_rate)
    summary_mutations_data_list.append(summary_data)



## Step 9: Create summary table
##--------------------

print("Step 9: Create summay table")

summary_df = pd.DataFrame(summary_mutations_data_list)
summary_excel_file = output_prefix+"summary_data.xlsx"
summary_df.to_excel(summary_excel_file, index=False)


## Step 10: Create summary graphs
##--------------------

print("Step 10: Create summary graphs (using a fixed threshold)")

for threshold in thresholds_list:
  error_rates_graph = plot_error_rate(output_prefix, summary_df, RT_error_rate, threshold=threshold)
  all_error_rates_graph = plot_all_error_rates(output_prefix, summary_df, RT_error_rate, threshold=threshold)
```

```python
#@markdown # Offsprings graph (optional)





## Functions
##----------

def find_offspring_barcodes(barcodes_dic):
    """
    Identify potential offspring barcodes based on nucleotide differences with higher-frequency barcodes.

    Parameters:
        barcodes_dic (dict): Dictionary containing barcodes as keys and their corresponding frequencies as values.

    Returns:
        tuple: A tuple containing two dictionaries:
            - offsprings_1differenc_dic (dict): Barcodes with 1 difference from other higher-frequency barcodes
            - offsprings_2differenc_dic (dict): Barcodes with 2 differences from other higher-frequency barcodes
    """
    global offsprings_1differenc_dic
    global offsprings_2differenc_dic
    global offsprings_3differenc_dic
    global offsprings_4differenc_dic

    offsprings_1differenc_dic = defaultdict(int)
    offsprings_2differenc_dic = defaultdict(int)
    offsprings_3differenc_dic = defaultdict(int)
    offsprings_4differenc_dic = defaultdict(int)
    # Iterate over all barcode sequences in the list
    for barcode1, barcode1_freq in tqdm(barcodes_dic.items()):
        for barcode2, barcode2_freq in barcodes_dic.items():
            if barcode1_freq < barcode2_freq:
                # Count nucleotide differences between barcode sequences barcode1 and barcode2, ignoring case
                differences_count = sum(1 for nt1, nt2 in zip(barcode1.lower(), barcode2.lower()) if nt1 != nt2)
                # If we find 1 or 2 differences, we consider that barcode1 is an offspring of barcode2
                if differences_count == 1:
                    offsprings_1differenc_dic[barcode1] = barcode1_freq
                elif differences_count == 2:
                    offsprings_2differenc_dic[barcode1] = barcode1_freq
                elif differences_count == 3:
                    offsprings_3differenc_dic[barcode1] = barcode1_freq
                elif differences_count == 4:
                    offsprings_4differenc_dic[barcode1] = barcode1_freq

    # Print summary statistics
    print("\t\tTotal barcodes: ", len(barcodes_dic))
    print("\t\tBarcodes with 1 difference from other barcodes with higher frequency: ", len(offsprings_1differenc_dic))
    print("\t\tBarcodes with 2 differences from other barcodes with higher frequency: ", len(offsprings_2differenc_dic))
    return offsprings_1differenc_dic, offsprings_2differenc_dic, offsprings_3differenc_dic, offsprings_4differenc_dic

def calculate_frequency_of_frequencies_using_barcodes_dic(barcodes_dic):
    """
    Calculate the frequencies of frequencies in the given dictionary.

    Parameters:
        barcodes_dic (dict): Dictionary containing barcodes as keys and their corresponding frequencies as values.

    Returns:
        sorted_result_dict (dict): A dictionary where keys are frequencies and values are the count of occurrences of each frequency.
    """
    frequencies = list(barcodes_dic.values())
    frequency_counter = Counter(frequencies)
    sorted_result_dict = dict(sorted(frequency_counter.items()))
    return sorted_result_dict


def save_combined_frequencies(counted_barcodes_frequencies_dic, counted_offsprings_1differenc_frequencies_dic, counted_offsprings_2differenc_frequencies_dic, counted_offsprings_3differenc_frequencies_dic, counted_offsprings_4differenc_frequencies_dic):
    # Combine all dictionaries into one
    combined_dic = {
        "barcodes": counted_barcodes_frequencies_dic,
        "offsprings_1difference": counted_offsprings_1differenc_frequencies_dic,
        "offsprings_2difference": counted_offsprings_2differenc_frequencies_dic,
        "offsprings_3difference": counted_offsprings_3differenc_frequencies_dic,
        "offsprings_4difference": counted_offsprings_4differenc_frequencies_dic
    }
    # Save the combined dictionary to a file
    offsprings_data_file = "offsprings_data.json"
    with open(offsprings_data_file, "w") as outfile:
        json.dump(combined_dic, outfile)

    # Return the name of the created file
    return offsprings_data_file


def plot_all_frequencies_percentages_graph(output_prefix, counted_barcodes_frequencies_dic, counted_offsprings_1differenc_frequencies_dic, counted_offsprings_2differenc_frequencies_dic,counted_offsprings_3differenc_frequencies_dic, counted_offsprings_4differenc_frequencies_dic, counted_offsprings_barcodes_frequencies_dic):
    """
    Plot the frequencies of frequencies for the counted_barcodes_frequencies_dic and the percentage of counted frequencies for counted_offsprings_1differenc_frequencies_dic and counted_offsprings_2differenc_frequencies_dic

    Parameters:
        counted_barcodes_frequencies_dic, counted_offsprings_1differenc_frequencies_dic, counted_offsprings_2differenc_frequencies_dic (dict): Dictionaries where keys are frequencies
                                                                                                                                               and values are the count of occurrences of each frequency.
                                                                                                                                               Dictionaries are assumed to be sorted by frequencies in ascending order.    """

    # Extract frequencies and occurrences/percentages for each dataset
    frequencies1, occurrences1 = list(counted_barcodes_frequencies_dic.keys()), list(counted_barcodes_frequencies_dic.values())

    percentage_offsprings_1differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_1differenc_frequencies_dic.items()}
    frequencies2, percentages2 = list(percentage_offsprings_1differenc_frequencies_dic.keys()), list(percentage_offsprings_1differenc_frequencies_dic.values())

    percentage_offsprings_2differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_2differenc_frequencies_dic.items()}
    frequencies3, percentages3 = list(percentage_offsprings_2differenc_frequencies_dic.keys()), list(percentage_offsprings_2differenc_frequencies_dic.values())

    percentage_offsprings_3differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_3differenc_frequencies_dic.items()}
    frequencies4, percentages4 = list(percentage_offsprings_3differenc_frequencies_dic.keys()), list(percentage_offsprings_3differenc_frequencies_dic.values())

    percentage_offsprings_4differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_4differenc_frequencies_dic.items()}
    frequencies5, percentages5 = list(percentage_offsprings_4differenc_frequencies_dic.keys()), list(percentage_offsprings_4differenc_frequencies_dic.values())

    percentage_offsprings_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_barcodes_frequencies_dic.items()}
    frequencies6, percentages6 = list(percentage_offsprings_frequencies_dic.keys()), list(percentage_offsprings_frequencies_dic.values())

    # Create a scatter plot
    fig, ax = plt.subplots()
    ax2 = ax.twinx()
    ax2.set_ylim(0, 101)
    ax.scatter(frequencies1, occurrences1, color='tab:blue', label='All barcodes frequencies', marker='o')
    ax2.plot(frequencies2, percentages2, 's-', color='tab:red', label='% Barcodes with 1 difference from other higher-frequency barcodes')
    ax2.plot(frequencies3, percentages3, '^-', color='tab:green', label='% Barcodes with 2 differences from other higher-frequency barcodes')
    ax2.plot(frequencies4, percentages4, '^-', color='tab:purple', label='% Barcodes with 3 differences from other higher-frequency barcodes')
    ax2.plot(frequencies5, percentages5, '^-', color='tab:brown', label='% Barcodes with 4 differences from other higher-frequency barcodes')
    ax2.plot(frequencies6, percentages6, '^-', color='tab:orange', label='% Real offsprings')

    # Add labels and title
    ax.set_xlabel('Barcodes frequencies')
    ax.set_ylabel('Occurrences')
    ax2.set_ylabel('% of offspring barcodes')
    ax.set_title('Offspring barcodes distribution')

    # Set x-axis to log scale
    plt.xscale('log')

    # Add legends
    ax.legend(loc='center left', bbox_to_anchor=(1.2, 0.75))
    ax2.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))

    # Save the plot
    plot_file_percentages  = output_prefix+"percentage_differences_all.png"
    plt.savefig(plot_file_percentages, bbox_inches='tight')
    plt.show()
    plt.close()
    return plot_file_percentages

def plot_frequencies_percentages_graph(output_prefix, counted_barcodes_frequencies_dic, counted_offsprings_1differenc_frequencies_dic, counted_offsprings_2differenc_frequencies_dic, counted_offsprings_barcodes_frequencies_dic):
    """
    Plot the frequencies of frequencies for the counted_barcodes_frequencies_dic and the percentage of counted frequencies for counted_offsprings_1differenc_frequencies_dic and counted_offsprings_2differenc_frequencies_dic

    Parameters:
        counted_barcodes_frequencies_dic, counted_offsprings_1differenc_frequencies_dic, counted_offsprings_2differenc_frequencies_dic (dict): Dictionaries where keys are frequencies
                                                                                                                                               and values are the count of occurrences of each frequency.
                                                                                                                                               Dictionaries are assumed to be sorted by frequencies in ascending order.    """
    # Extract frequencies and occurrences/percentages for each dataset
    frequencies1, occurrences1 = list(counted_barcodes_frequencies_dic.keys()), list(counted_barcodes_frequencies_dic.values())

    percentage_offsprings_1differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_1differenc_frequencies_dic.items()}
    frequencies2, percentages2 = list(percentage_offsprings_1differenc_frequencies_dic.keys()), list(percentage_offsprings_1differenc_frequencies_dic.values())

    percentage_offsprings_2differenc_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_2differenc_frequencies_dic.items()}
    frequencies3, percentages3 = list(percentage_offsprings_2differenc_frequencies_dic.keys()), list(percentage_offsprings_2differenc_frequencies_dic.values())

    percentage_offsprings_frequencies_dic = {frequence: occurrence / counted_barcodes_frequencies_dic[frequence] * 100 for frequence, occurrence in counted_offsprings_barcodes_frequencies_dic.items()}
    frequencies6, percentages6 = list(percentage_offsprings_frequencies_dic.keys()), list(percentage_offsprings_frequencies_dic.values())

    # Create a scatter plot
    fig, ax = plt.subplots()
    ax2 = ax.twinx()
    ax2.set_ylim(0, 101)
    ax.scatter(frequencies1, occurrences1, color='tab:blue', label='All barcodes frequencies', marker='o')
    ax2.plot(frequencies2, percentages2, 's-', color='tab:red', label='% Barcodes with 1 difference from other higher-frequency barcodes')
    ax2.plot(frequencies3, percentages3, '^-', color='tab:green', label='% Barcodes with 2 differences from other higher-frequency barcodes')
    ax2.plot(frequencies6, percentages6, '^-', color='tab:orange', label='% Real offsprings')

    # Add labels and title
    ax.set_xlabel('Barcodes frequencies')
    ax.set_ylabel('Occurrences')
    ax2.set_ylabel('% of offspring barcodes')
    ax.set_title('Offspring barcodes distribution')

    # Set x-axis to log scale
    plt.xscale('log')

    # Add legends
    ax.legend(loc='center left', bbox_to_anchor=(1.2, 0.75))
    ax2.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))

    # Save the plot
    plot_file_percentages  = output_prefix+"percentage_differences.png"
    plt.savefig(plot_file_percentages, bbox_inches='tight')
    plt.show()
    plt.close()
    return plot_file_percentages






## Main program
##-------------

## Step 11: Finding offspring barcodes
##--------------------

print("Step 11: Finding offspring barcodes")

offsprings_1differenc_dic, offsprings_2differenc_dic, offsprings_3differenc_dic, offsprings_4differenc_dic = find_offspring_barcodes(barcodes_dic)



## Step 12: Calculating number of occurrences for offspring barcodes frequencies
##--------------------

print("Step 12 => Calculating number of occurrences for offspring barcodes frequencies ")

counted_barcodes_frequencies_dic = calculate_frequency_of_frequencies_using_barcodes_dic(barcodes_dic)
counted_offsprings_1differenc_frequencies_dic = calculate_frequency_of_frequencies_using_barcodes_dic(offsprings_1differenc_dic)
counted_offsprings_2differenc_frequencies_dic = calculate_frequency_of_frequencies_using_barcodes_dic(offsprings_2differenc_dic)
counted_offsprings_3differenc_frequencies_dic = calculate_frequency_of_frequencies_using_barcodes_dic(offsprings_3differenc_dic)
counted_offsprings_4differenc_frequencies_dic = calculate_frequency_of_frequencies_using_barcodes_dic(offsprings_4differenc_dic)

offsprings_data_file = save_parameters_to_file(output_prefix+"offsprings_data.json",
                        counted_barcodes_frequencies_dic=counted_barcodes_frequencies_dic,
                        counted_offsprings_1differenc_frequencies_dic=counted_offsprings_1differenc_frequencies_dic,
                        counted_offsprings_2differenc_frequencies_dic=counted_offsprings_2differenc_frequencies_dic,
                        counted_offsprings_3differenc_frequencies_dic=counted_offsprings_3differenc_frequencies_dic,
                        counted_offsprings_4differenc_frequencies_dic=counted_offsprings_4differenc_frequencies_dic)

print("\t\tOccurrences of each frequency for all barcodes: ", counted_barcodes_frequencies_dic)
print("\t\tOccurrences of each frequency for barcodes with 1 difference from other barcodes with higher frequency: ", counted_offsprings_1differenc_frequencies_dic)
print("\t\tOccurrences of each frequency for barcodes with 2 differences from other barcodes with higher frequency: ", counted_offsprings_2differenc_frequencies_dic)
print("\t\tOccurrences of each frequency for barcodes with 3 differences from other barcodes with higher frequency: ", counted_offsprings_3differenc_frequencies_dic)
print("\t\tOccurrences of each frequency for barcodes with 4 differences from other barcodes with higher frequency: ", counted_offsprings_4differenc_frequencies_dic)



## Step 13: Calculating number of occurrences for offspring barcodes frequencies
##--------------------

print("Step 13: Calculating number of occurrences for offspring barcodes frequencies")

all_offspring_percentages_graph = plot_all_frequencies_percentages_graph(output_prefix,
                                                                         counted_barcodes_frequencies_dic,
                                                                         counted_offsprings_1differenc_frequencies_dic,
                                                                         counted_offsprings_2differenc_frequencies_dic,
                                                                         counted_offsprings_3differenc_frequencies_dic,
                                                                         counted_offsprings_4differenc_frequencies_dic,
                                                                         counted_offsprings_barcodes_frequencies_dic)
offspring_percentages_graph = plot_frequencies_percentages_graph(output_prefix,
                                                                 counted_barcodes_frequencies_dic,
                                                                 counted_offsprings_1differenc_frequencies_dic,
                                                                 counted_offsprings_2differenc_frequencies_dic,
                                                                 counted_offsprings_barcodes_frequencies_dic)




print("#############################")
print("END")
print("#############################")
```
